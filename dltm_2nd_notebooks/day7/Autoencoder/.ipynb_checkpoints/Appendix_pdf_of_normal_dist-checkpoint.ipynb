```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mu = 0
std = np.sqrt(20)
rv = norm(mu, std)

x = np.linspace(-15, 15, 1000)
plt.plot(x, rv.pdf(x))
plt.ylabel("p(x)")
plt.title("pdf of normal distribution with mean=0 and var=20")
plt.show()
```

```python
mu = 0
std = 1
rv = norm(mu, std)

x = np.linspace(-15, 15, 1000)
plt.plot(x, rv.pdf(x))
plt.ylabel("p(x)")
plt.title("pdf of normal distribution with mean=0 and var=1")
plt.show()
```
