```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.misc import derivative
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def tanh(x):
    return (np.exp(2 * x) - 1) / (np.exp(2 * x) + 1) 
```

```python
def relu(x):
    if x >= 0:
        return x
    else:
        return 0
```

```python
def relu(x):
    return max(0, x)
```

```python
test = np.linspace(-5, 5, 1000)
```

```python
test_sigmoid = [sigmoid(x) for x in test]
test_tanh = [tanh(x) for x in test]
test_relu = [relu(x) for x in test]
```

```python
plt.plot(test, test_sigmoid, 'r-', label='sigmoid')
plt.plot(test, test_tanh, 'b-', label='tanh')
plt.plot(test, test_relu, 'g-', label='relu')
axes = plt.gca()
axes.set_ylim([-1.5, 1.5])
plt.legend()
plt.show()
```

```python
def sigmoid_dv(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
def tanh_dv(x):
    return 1 - np.power(tanh(x), 2)
```

```python
def relu_dv(x):
    if x > 0: return 1
    elif x <= 0: return 0
```

```python
test_sigmoid_dv = [sigmoid_dv(x) for x in test]
test_tanh_dv = [tanh_dv(x) for x in test]
test_relu_dv = [relu_dv(x) for x in test]
```

```python
plt.plot(test, test_sigmoid_dv, 'r-', label='derivative of sigmoid')
plt.plot(test, test_tanh_dv, 'b-', label='derivative of tanh')
plt.plot(test, test_relu_dv, 'g-', label='derivative of relu')
axes = plt.gca()
axes.set_ylim([-0.01, 1.1])
plt.legend()
plt.show()
```

```python
plt.plot(test, test_sigmoid, 'k-', label='sigmoid')
plt.plot(test, test_sigmoid_dv, 'r--', label='derivative of sigmoid')
axes = plt.gca()
axes.set_ylim([-0.01, 1.1])
plt.legend()
plt.show()
```

```python
plt.plot(test, test_tanh, 'k-', label='tanh')
plt.plot(test, test_tanh_dv, 'r--', label='derivative of tanh')
axes = plt.gca()
axes.set_ylim([-1.1, 1.1])
plt.legend()
plt.show()
```

```python
plt.plot(test, test_relu, 'k-', label='relu')
plt.plot(test, test_relu_dv, 'r--', label='derivative of relu')
axes = plt.gca()
axes.set_ylim([-0.1, 3])
plt.legend()
plt.show()
```

```python
def prelu(x, alpha=0.1):
    if x >= 0:
        return x
    else:
        return alpha * x
```

```python
plt.plot(test, test_relu, 'b-', label='relu')
axes = plt.gca()
axes.set_ylim([-1, 3])
plt.legend()
plt.show()
```

```python
test_prelu = [prelu(x) for x in test]
```

```python
plt.plot(test, test_prelu, 'b-', label='prelu')
axes = plt.gca()
axes.set_ylim([-1, 3])
plt.legend()
plt.show()
```
